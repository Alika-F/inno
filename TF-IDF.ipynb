```python
import nltk
nltk.download('punkt')
import string
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from pymorphy2 import MorphAnalyzer
```

```python
nltk.download('stopwords')
russian_stopwords = set(stopwords.words('russian'))
morph = MorphAnalyzer()
```

```python
# Тексты песен
songs = {
    "В городе лето": """
    И время мимо так молчаливо над крышами города
    Кружит печально какой-то случайный вальс
    И час до полночи, телефон молчит тихо и холодно
    Я так скучаю, мне так не хватает Вас

    А небо всё так же высоко
    И солнце по крышам, и в городе лето
    И вся моя жизнь одной строкой
    Меня ты не слышишь, звонки без ответа

    Иду по улицам, город хмурится серым лицом
    И сыпет словами, увы, бесполезными
    Мне надо забыть её, надо забыть, и дело с концом
    Как будто бы мне полсердца отрезали!
    """,
    "Кажется": """
    Я в поездах, отставших от расписанья
    Я в городах, которых нету на карте
    Я умер, так и не пришедши в сознанье
    Хотел быть первым, но споткнулся на старте

    Я помню чёрные края этой бездны
    Я помню пение живущих там вечно
    Я понял, что слова теперь бесполезны
    Я видел небо - оно бесконечно

    Кажется, никогда нам на планете этой не встретиться
    Просто рукой махнуть - и голос почти не слышен
    Просто "пока-пока", и молча пойду я вниз по лестнице
    Может, когда - нибудь музыка станет тише
    """,
    "Проститься": """
    По минутам
    Осыпается
    Ожидание невозможного
    Ранним утром просыпается
    От движения неосторожного
    Как молчание ледяной зимы
    Нас закутало неизвестностью
    Здесь так долго друг друга
    Искали мы И конечно
    Пропали без вести

    Проститься
    Нету сил закрываю
    Я глаза закрываю
    Сквозь туман уплывая
    По аллеям столицы
    Проститься
    За потерей потеря
    И года полетели
    За дождями метели
    Перелетные птицы
    """
}
```

```python
#лемматизация текста и удаление стоп-слов
def preprocess_text(text):
    words = nltk.word_tokenize(text.lower())
    lemmatized_words = [morph.parse(word)[0].normal_form for word in words if word not in russian_stopwords and word not in string.punctuation]
    return ' '.join(lemmatized_words)

processed_songs = {title: preprocess_text(lyrics) for title, lyrics in songs.items()}
```

```python
# Инициализация векторизатора TF-IDF
tfidf_vectorizer = TfidfVectorizer()

# Преобразование текстов в матрицу TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(processed_songs.values())

feature_names = tfidf_vectorizer.get_feature_names_out()
```

```python
import pandas as pd

tfidf_df = pd.DataFrame(tfidf_matrix.T.toarray(), index=feature_names, columns=processed_songs.keys())
pd.set_option('display.max_rows', None)
tfidf_df
```
