```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
data = pd.read_csv('/content/diabetes.csv')
```

```python
data
```

k-means для показателя BMI,возраста Age

```python
X=data[['BMI','Age']]
y=data['Outcome']
```

```python
#изначальная выборка
plt.figure(figsize=(15, 8))
plt.scatter(X['BMI'][y==0], X['Age'][y==0], c='blue', label='no diabet')
plt.scatter(X['BMI'][y==1], X['Age'][y==1], c='red', label='diabet')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.legend(fontsize=12)
plt.show()
```

```python
kmeans = KMeans(n_clusters=2,random_state=42)
y_pred = kmeans.fit_predict(X)
# метрика
accuracy=np.sum(y_pred==y)/ len(y_pred)
print("Точность:",accuracy)
```

```python
plt.figure(figsize=(15, 8))
plt.scatter(X['BMI'][y_pred==0], X['Age'][y_pred==0], c='blue', label='no diabet')
plt.scatter(X['BMI'][y_pred==1], X['Age'][y_pred==1], c='red', label='diabet')
plt.xlabel('BMI')
plt.ylabel('Age')
plt.legend(fontsize=12)
plt.show()
```

Метод иерархической кластеризации с выводом дендрограммы для показателя BMI, возраста Age

```python
X=data[['BMI','Age']]
y=data['Outcome']
```

```python
# Определяем вывод
fig = plt.figure(figsize=(16, 50))
fig.patch.set_facecolor('white')

# Реализация иерархической кластеризации при помощи функции linkage
mergings = linkage(X, method='ward')

# Построение дендрограммы. Разными цветами выделены автоматически определенные кластеры
R = dendrogram(mergings, labels=[y[i] for i in y], orientation = 'left', leaf_font_size = 4)

# Отображение дендрограммы
plt.show()
```
