```python
#импортируем библиотеки
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
```

```python
# загружаем набор данных
data = pd.read_csv('/content/diabetes.csv')
data
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).reshape(-1, 1)

X
```

3 слоя

```python
# определяем модель с 3 внутренними слоями
model_3_layers = nn.Sequential(
    nn.Linear(8, 16),
    nn.ReLU(),
    nn.Linear(16, 16),
    nn.ReLU(),
    nn.Linear(16, 16),
    nn.ReLU(),
    nn.Linear(16, 8),
    nn.ReLU(),
    nn.Linear(8, 1),
    nn.Sigmoid()
)
```

```python
# тренируем модель с 3 внутренними слоями
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model_3_layers.parameters(), lr=0.001)
```

```python
n_epochs = 100
batch_size = 10

for epoch in range(n_epochs):
    for i in range(0, len(X), batch_size):
        Xbatch = X[i:i+batch_size]
        y_pred = model_3_layers(Xbatch)
        ybatch = y[i:i+batch_size]
        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Итерация № {epoch}, latest loss {loss}')
```

```python
# высчитываем accuracy для модели с 3 внутренними слоями
with torch.no_grad():
    y_pred = model_3_layers(X)
accuracy_3_layers = (y_pred.round() == y).float().mean()
print(f"Accuracy модели с 3 внутренними слоями: {accuracy_3_layers}")
```

4 слоя

```python
# модель с 4 внутренними слоями
model_4_layers = nn.Sequential(
    nn.Linear(8, 16),
    nn.ReLU(),
    nn.Linear(16, 16),
    nn.ReLU(),
    nn.Linear(16, 16),
    nn.ReLU(),
    nn.Linear(16, 16),
    nn.ReLU(),
    nn.Linear(16, 8),
    nn.ReLU(),
    nn.Linear(8, 1),
    nn.Sigmoid()
)
```

```python
# тренируем модель с 4 внутренними слоями
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model_4_layers.parameters(), lr=0.001)
```

```python
for epoch in range(n_epochs):
    for i in range(0, len(X), batch_size):
        Xbatch = X[i:i+batch_size]
        y_pred = model_4_layers(Xbatch)
        ybatch = y[i:i+batch_size]
        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Итерация № {epoch}, latest loss {loss}')
```

```python
# высчитываем accuracy для модели с 4 внутренними слоями
with torch.no_grad():
    y_pred = model_4_layers(X)
accuracy_4_layers = (y_pred.round() == y).float().mean()
print(f"Accuracy модели с 4 внутренними слоями: {accuracy_4_layers}")
```
